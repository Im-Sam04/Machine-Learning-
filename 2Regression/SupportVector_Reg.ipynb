```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X)
```

```python
print(y)
```

```python
y = y.reshape(len(y), 1)
print(y)
```

```python
from sklearn.preprocessing import StandardScaler

sc_X= StandardScaler()
X = sc_X.fit_transform(X)
sc_y= StandardScaler()
y = sc_y.fit_transform(y)
print(X)
print(y)
```

```python
from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf')
regressor.fit(X ,y)
```

```python
sc_y.inverse_transform(regressor.predict(sc_X.fit_transform([[6.5]])).reshape(-1, 1))
```

```python
plt.scatter(sc_X.inverse_transform(X), sc_y.inverse_transform(y), color="red")
plt.plot(sc_X.inverse_transform(X),sc_y.inverse_transform(regressor.predict(X).reshape(-1, 1)) , color="blue")
plt.title("True or Bluff (SVR)")
plt.xlabel(" Position Level")
plt.ylabel("Salary")
plt.show()
```

```python

X_grid = np.arange(min(sc_X.inverse_transform(X)), max(sc_X.inverse_transform(X)), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))  # Reshape to 2D for compatibility

plt.scatter(sc_X.inverse_transform(X), sc_y.inverse_transform(y), color='red')


plt.plot(
    X_grid,
    sc_y.inverse_transform(regressor.predict(sc_X.transform(X_grid)).reshape(-1, 1)),  # Reshape predictions
    color='blue'
)


plt.title("True or Bluff (SVR)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```
